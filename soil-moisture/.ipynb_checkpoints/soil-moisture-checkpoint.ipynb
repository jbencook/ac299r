```python
%matplotlib inline
%load_ext rmagic
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import datetime as dt
from scipy.spatial import distance_matrix
from operator import itemgetter
import scipy.spatial as spatial
import psycopg2
import os
import psycopg2.extras as extras
import seaborn
import warnings

warnings.filterwarnings("ignore")
```

### Connect to the database

```python
con = psycopg2.connect(database=os.environ['RDS_DB_NAME'], 
                       password=os.environ['RDS_PASSWORD'],
                       user=os.environ['RDS_USERNAME'],
                       host=os.environ['RDS_HOSTNAME'])
cur = con.cursor(cursor_factory=extras.DictCursor)
```

### Get all station IDs

```python
cur.execute("select distinct(site_id) from nwcc;")
res = cur.fetchall()
```

```python
sites = map(itemgetter(0), res)
```

### For each station, get all observations (precipitaiton and soil moisture) sorted by date

```python
data = {}

for i,s in enumerate(sites):
    cur.execute("select prcp, sms from nwcc where site_id=%s order by date;" % s)
    res = cur.fetchall()
    data[s] = np.array(res)
```

```python
def mc(x):
    """Subtract the mean and divide by the standard deviation
    """
    return (x - x.mean())/x.std()
```

## Plot precipitation (blue) and soil moisture (red) to see the relationship

```python
i = 1

fig = plt.figure(figsize=(20,5))


keys = data.keys()
k = sites[0]

prcp = data[k][:,0]
sms = data[k][:,1]

plt.bar(np.arange(len(prcp)), mc(prcp))
plt.plot(mc(sms), c="red", lw=2)
fig.savefig("county1.png")
```

### Create matrices

```python
num_obs = 0
for k in data:
    num_obs += (data[k].shape[0]-1)
```

```python
mt = np.zeros(num_obs)
mtminus1 = np.zeros((num_obs, len(sites)))
pt = np.zeros((num_obs, len(sites)))
pt_m1 = np.zeros((num_obs, len(sites)))
```

```python
cur_begin = 0
ndx_site = {}
site_ndx = {}
for i,k in enumerate(data):
    ndx_site[i] = k
    site_ndx[k] = i
    n_s = data[k].shape[0] - 1
    mt[cur_begin:(cur_begin + n_s)] = data[k][1:,1]
    mtminus1[cur_begin:(cur_begin + n_s),i] = data[k][:-1,1]
    #pt[cur_begin:(cur_begin + n_s),i] = data[k][1:,0]
    pt[cur_begin:(cur_begin + n_s),i] = data[k][:-1,0]
    
    cur_begin += n_s
```

### Censor

```python
mt[mt < 0] = 0
mtminus1[mtminus1 < 0] = 0
pt[pt < 0] = 0
# pt[pt > 200] = 200
# pt_m1[pt_m1 < 0] = 0
# pt_m1[pt_m1 > 200] = 200
```

```python
X = np.column_stack((mtminus1, pt))
Xp = np.column_stack((np.ones(len(mt)), X))
```

### Run least squares

```python
B = np.linalg.pinv(Xp.T.dot(Xp)).dot(Xp.T.dot(mt))
```

### Get coefficients

```python
alpha, beta = B[1:(len(sites)+1)], B[(len(sites) + 1):]
```

### $\alpha_s$ for least squares

```python
plt.figure(figsize=(7,7))
plt.hist(alpha, bins=25)
plt.title("a_s")
plt.show()
```

### $\beta_s$ for least squares

```python
plt.figure(figsize=(7,7))
plt.hist(beta, bins=25)
plt.title("b_s")
plt.show()
```

### Get all the station data and put it in a dataframe

```python
def fetch_df(query):
    cur.execute(query)
    res = cur.fetchall()
    try:
        cols = res[0].keys()
        df = {}
        for c in cols:
            df[c] = map(itemgetter(c), res)
        return pd.DataFrame(df)
    except:
        return None
```

```python
nwcc_station = fetch_df("select * from nwcc_station;")
```

### Create distance matrix

```python
s = np.empty((len(sites), 2))
for i in xrange(len(sites)):
    s[i,:] = nwcc_station[nwcc_station['id'] == ndx_site[i]][['lon', 'lat']].irow(0)
    
def normalize_01(x):
    return (x - x.min()) / (x.max() - x.min())

s[:,0] = normalize_01(s[:,0])
s[:,1] = normalize_01(s[:,1])
    
d = spatial.distance_matrix(s, s)
```

### Stations we observe

```python
plt.plot(s[:,0], s[:,1], 'o')
```

### Ridge regression

```python
def squared_exponential(d, sigma, phi):
    """Squared exponential covariance function
    """
    return sigma * np.exp( -np.square(d) / phi)

C = squared_exponential(d, 100, 10)
Z = np.zeros(C.shape)
```

```python
top = np.column_stack((C, Z))
bottom = np.column_stack((Z,C))
sigma = np.row_stack((top, bottom))
sigp = np.eye(2*len(sites) + 1)
sigp[1:,1:] = sigma
sigp_inv = np.linalg.pinv(sigp)
```

```python
B2 = np.linalg.pinv(Xp.T.dot(Xp) + sigp_inv).dot(Xp.T.dot(mt))
```

```python
alpha2, beta2 = B2[1:(len(sites)+1)], B2[(len(sites)+1):]
```

```python
plt.figure(figsize=(7,7))
plt.hist(alpha2, bins=25)
plt.title("a_s")
plt.show()
```

```python
plt.figure(figsize=(7,7))
plt.hist(beta2, bins=25)
plt.title("b_s")
plt.show()
```

### $R^2$ for least squares

```python
1 - np.sum(np.square(mt - Xp.dot(B))) / np.sum(np.square(mt - mt.mean()))
```

### $R^2$ for ridge regression

```python
1 - np.sum(np.square(mt - Xp.dot(B2))) / np.sum(np.square(mt - mt.mean()))
```

### Predicted vs. Observed for Ridge Regression

```python
plt.plot(mt, Xp.dot(B2), '.', alpha=.25)
plt.xlim([0,100])
plt.ylim([0,100])
plt.show()
```

```python
Xp_mtminus = Xp[:,:(len(sites)+1)]
B_mtminus1 = np.linalg.pinv(Xp_mtminus.T.dot(Xp_mtminus)).dot(Xp_mtminus.T.dot(mt))
Xp_pt = np.column_stack((np.ones(Xp.shape[0]),Xp[:,(len(sites)+1)]))
B_pt = np.linalg.pinv(Xp_pt.T.dot(Xp_pt)).dot(Xp_pt.T.dot(mt))
```

### $R^2$ for $M_{t-1}$ only

```python
1 - np.sum(np.square(mt - Xp_mtminus.dot(B_mtminus1))) / np.sum(np.square(mt - mt.mean()))
```

### $R^2$ for $P_{t}$ only

```python
1 - np.sum(np.square(mt - Xp_pt.dot(B_pt))) / np.sum(np.square(mt - mt.mean()))
```
