# Data

The data for this project come from a variety of sources and are of varying quality. 
All of the data are freely available on the web through web scraping, automatic programming interfaces (APIs), or by request.

One of the most important variables is estimated average maize yield (in bushels per acre) for a given county in a given year. 
These data are collected with sample surveys by the National Agricultural Statistics Service (NASS) of the United States Department of Agriculture (USDA).
I download these data automatically using the NASS Quick Stats API.

Climate data come from two sources: the United States Historical Climatology Network (USHCN) and the Natural Resources Conservation Service (NRCS).
There are roughly 1,200 USHCN weather stations recording daily precipitation (in hundredths of an inch), daily minimum temperature (in degrees fahrenheit) and daily maximum temperature (also in degrees fahrenheit).
Temperature is typically summarized by Growing Degree Days (GDD), which is a function of the daily minimum and maximum temperatures.
Soil moisture comes from roughly 600 NRCS stations and is reported as a percentage.

Finally, I include variables about the soil itself as covariates in some of the models.
It is possible to scrape a plethora of information about the soil at any given point in the United States from a website maintained by the California Resource Soil Lab at the University of California, Davis.
Possible relevant variables include available water storage (0 - 100cm), drainage class (a categorical variable indicating how well drained the soil is), and proportions of sand, silt and clay.
Because these data are defined as polygons and can therefore be queried at any given point in the United States, including these data will also require some kind of spatial dimensionality reduction.

After storing Amazon RDS variables in the enviornment, we can register the PostGIS database with `psycopg2`.

```python
print os.environ['RDS_DB_NAME']
```

```python
con = psycopg2.connect(database=os.environ['RDS_DB_NAME'], 
                       password=os.environ['RDS_PASSWORD'],
                       user=os.environ['RDS_USERNAME'],
                       host=os.environ['RDS_HOSTNAME'])
cur = con.cursor(cursor_factory=extras.DictCursor)
```

I define a function that takes a SQL query and returns a pandas dataframe with the result.

```python
def fetch_df(query):
    """
    Takes a PostgreSQL and returns the table in a pandas dataframe
    """
    cur.execute(query)
    res = cur.fetchall()
    try:
        cols = res[0].keys()
        df = {}
        for c in cols:
            df[c] = map(itemgetter(c), res)
        return pd.DataFrame(df)
    except:
        return None
```

Now it's possible to make GIS queries on the data. Below, I get the centroid of all the counties in Nebraska.

```python
nebraska_counties = fetch_df("select c.name, st_x(st_centroid(c.geom)) as lon, " + 
                             "st_y(st_centroid(c.geom)) as lat from county as c, state as s " + 
                             "where c.state_id=s.id and s.statefp='31';")
```

And I find the average yield in bushels per acre over time.

```python
average_yield = fetch_df("select year, avg(yield) as yield from area_yield group by year order by year;")
```

```python
plt.plot(average_yield['year'], average_yield['yield'], 'o')
plt.xlabel("year")
plt.ylabel("Bushels per acre")
plt.show()
```
