```python
%matplotlib inline
import os

import brewer2mpl
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.decomposition import pca
from sklearn.linear_model import LinearRegression

import warnings
warnings.filterwarnings('ignore')

# Set up some better defaults for matplotlib
from matplotlib import rcParams

#colorbrewer2 Dark2 qualitative color table
dark2_colors = brewer2mpl.get_map('Dark2', 'Qualitative', 7).mpl_colors

rcParams['figure.figsize'] = (10, 6)
rcParams['figure.dpi'] = 150
rcParams['axes.color_cycle'] = dark2_colors
rcParams['lines.linewidth'] = 2
rcParams['axes.facecolor'] = 'white'
rcParams['font.size'] = 14
rcParams['patch.edgecolor'] = 'white'
rcParams['patch.facecolor'] = dark2_colors[0]
rcParams['font.family'] = 'StixGeneral'


def remove_border(axes=None, top=False, right=False, left=True, bottom=True):
    """
    Minimize chartjunk by stripping out unnecesasry plot borders and axis ticks
    
    The top/right/left/bottom keywords toggle whether the corresponding plot border is drawn
    """
    ax = axes or plt.gca()
    ax.spines['top'].set_visible(top)
    ax.spines['right'].set_visible(right)
    ax.spines['left'].set_visible(left)
    ax.spines['bottom'].set_visible(bottom)
    
    #turn off all ticks
    ax.yaxis.set_ticks_position('none')
    ax.xaxis.set_ticks_position('none')
    
    #now re-enable visibles
    if top:
        ax.xaxis.tick_top()
    if bottom:
        ax.xaxis.tick_bottom()
    if left:
        ax.yaxis.tick_left()
    if right:
        ax.yaxis.tick_right()
```

```python
# Load data
prcp = pd.read_csv(os.path.join(os.getcwd(), '../data/prcp.csv'))
gdd = pd.read_csv(os.path.join(os.getcwd(), '../data/gdd_c.csv'))
```

```python
# Filter data
pmat = prcp[prcp.columns[2:]]
gmat = gdd[gdd.columns[2:]]

pmat = np.array(pmat)[:,90:304] # Get April 1 - October 31
gmat = np.array(gmat)[:,90:304]
```

```python
# Center and scale
pmean = np.array([pmat.mean(axis=0),]*len(pmat))
pstd = np.array([pmat.std(axis=0),]*len(pmat))

gmean = np.array([gmat.mean(axis=0),]*len(gmat))
gstd = np.array([gmat.std(axis=0),]*len(gmat))

pmat = (pmat - pmean) / pstd
gmat = (gmat - gmean) / gstd
```

```python
# PCA
pca_obj = pca.PCA(n_components=5)

pmat_new = pca_obj.fit_transform(pmat)
gmat_new = pca_obj.fit_transform(gmat)
```

```python
# Create one DataFrame:
df = prcp[['station_code', 'year']]

for i in xrange(1,6):
    prcp_col = 'prcp_' + str(i)
    gdd_col = 'gdd_' + str(i)
    
    df[prcp_col] = pmat_new[:,i-1]
    df[gdd_col] = gmat_new[:,i-1]
```

```python
df.head()
```

```python
# Load in more data
cty_yield = pd.read_csv(os.path.join(os.getcwd(), '../data/area_yield.csv'))
cty_yield.irrigated = [1 if elem == 't' else 0 for elem in cty_yield.irrigated] # Convert 't' or 'f' to 1 or 0
cty_yield = cty_yield[(cty_yield.commodity_abbr == 'CN')] # Keep the Corn data
cty_yield.head()
```

```python
nearest_station = pd.read_csv(os.path.join(os.getcwd(), '../data/closest_station.csv'))
station_id_code = pd.read_csv(os.path.join(os.getcwd(), '../data/station_id_code.csv'))

cty_yield = pd.merge(cty_yield, nearest_station, how='left', on='county_id')
cty_yield = pd.merge(cty_yield, station_id_code, how='left', on='station_id')
```

```python
cty_yield.head()
```

```python
# This is equivalent to a SQL JOIN statement or merge in R
df2 = pd.merge(cty_yield, df, how='inner', on=['station_code', 'year'])
```

## Average corn yield over time in bushels per acre 

###Note: non-irrigated points are plotted $\frac{1}{2}$ year to the right in order to show where the distributions overlap.

```python
fig = plt.figure()
plt.scatter(df2.year[df2.irrigated==0] + .5, df2.yield_bu_ac[df2.irrigated==0], c=dark2_colors[1])
plt.scatter(df2.year[df2.irrigated==1], df2.yield_bu_ac[df2.irrigated==1], c=dark2_colors[0])
plt.title('Corn yield over time')
plt.ylabel('Yield (bu / ac)')
plt.xlabel('Year')
leg = plt.legend(('Non-Irrigated', 'Irrigated'), loc='upper left')
leg.draw_frame(False)
remove_border()
plt.show()
```

```python
# Make 2000 year 0
df2.year -= 2000
```

```python
# Define an interaction term
df2['irrigatedXyear'] = df2.irrigated * (df2.year)
```

```python
features = ['irrigated', 'year', 'irrigatedXyear',
            'prcp_1', 'prcp_2', 'prcp_3', 'prcp_4', 'prcp_5', 
            'gdd_1', 'gdd_2', 'gdd_3', 'gdd_4', 'gdd_5']
```

```python
y = np.array(df2.yield_bu_ac)
X = np.array(df2[features])

lm = LinearRegression()
lm.fit(X, y)
lm.coef_
```

```python
for i, f in enumerate(features):
    print '%14s: b = %5.2f' % (f, lm.coef_[i])
```

```python
y_hat = lm.predict(X)
```

##The following is a plot of predicted yield, $\hat{y}$, vs. observed yield, $y$, in bushels per acre.

```python
fig = plt.figure()
plt.scatter(y, y_hat, c=dark2_colors[1])
plt.title('$\hat{y}$ vs. $y$')
plt.ylabel('$\hat{y}$ (Estimated)')
plt.xlabel('$y$ (Observed)')
remove_border()
plt.show()
```

##The distribution of the residuals:

```python
residuals = y_hat - y
```

```python
fig = plt.figure()
plt.hist(residuals, normed=True, bins=25)
plt.title('Residuals')
plt.xlabel('$\hat{y} - y$')
plt.ylabel('Relative Frequency')
remove_border()
plt.show()
```

## The residuals over time

```python
fig = plt.figure()
plt.scatter(df2.year, residuals, c=dark2_colors[0])
plt.title('Residuals over time')
plt.ylabel('$\hat{y} - y$')
plt.xlabel('Year')
remove_border()
plt.show()
```

$R^2$

```python
print 'R^2 = %.3f' % lm.score(X, y)
```
