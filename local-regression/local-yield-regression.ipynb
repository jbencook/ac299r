```python
def remove_ticks():
    plt.xticks([])
    plt.yticks([])
    
def remove_border(axes=None, top=False, right=False, left=False, bottom=False):
    """
    Minimize chartjunk by stripping out unnecesasry plot borders and axis ticks
    The top/right/left/bottom keywords toggle whether the corresponding plot border is drawn
    """
    ax = axes or plt.gca()
    ax.spines['top'].set_visible(top)
    ax.spines['right'].set_visible(right)
    ax.spines['left'].set_visible(left)
    ax.spines['bottom'].set_visible(bottom)
    
    #turn off all ticks
    ax.yaxis.set_ticks_position('none')
    ax.xaxis.set_ticks_position('none')
    
    #now re-enable visibles
    if top:
        ax.xaxis.tick_top()
    if bottom:
        ax.xaxis.tick_bottom()
    if left:
        ax.yaxis.tick_left()
    if right:
        ax.yaxis.tick_right()
```

## Define `Polygon` that contains the field

```python
pts = np.array([[325,   9],
                [325, 670],
                [360, 670],
                [360, 610],
                [415, 610],
                [415, 670],
                [765, 670],
                [765,   9]], dtype=np.int)

pts[:,0] -= 325
pts[:,1] -= 9
```

```python
field = Polygon(pts)
```

## White out non-field zone

```python
num = 0
for i in xrange(img.shape[0]):
    for j in xrange(img.shape[1]):
        pt = Point(j,i) # shapely.geometry.Point object
        # if the point is part of the field, leave it alone, else make white
        if not field.contains(pt):
            img[i,j,:]  = 0
            img2[i,j,:] = 0
            img3[i,j,:] = 0
```

## Split field into zones

```python
num_x = 25 #Number of grid points in x direction
num_y = 50 #Number of grid points in y direction

xmin,ymin,xmax,ymax = field.bounds

x_pts = np.linspace(xmin, xmax, num_x)
y_pts = np.linspace(ymin, ymax, num_y)
```

```python
zones = []
for j in xrange(num_x - 1):
    for i in xrange(num_y - 1):
        z = box(x_pts[j], y_pts[i], x_pts[j+1], y_pts[i+1])
        if field.contains(z):
            z = field.intersection(z)
            zones.append(z)
```

```python
def plot_zone(z):
    pts = np.asarray(z.exterior.xy)
    plt.plot(pts[0,:],pts[1,:],'b')
```

## Plot the decomposed fields

```python
fig = plt.figure(figsize=(15,7))
plt.subplot(131)
plt.imshow(img2)
plt.xlabel('NDVI July 22')
remove_ticks()
remove_border()
[plot_zone(z) for z in zones]
plt.subplot(132)
plt.imshow(img3)
remove_ticks()
remove_border()
[plot_zone(z) for z in zones]
plt.xlabel('NDVI August 25')
plt.subplot(133)
plt.imshow(img)
remove_ticks()
remove_border()
[plot_zone(z) for z in zones]
plt.xlabel('Yield Map')
fig.savefig('decomposed_fields.png')
```

## Get yield values from colors

This is a naive method of getting yield values from the colors. Unfortunately, the values of the color channels in the gradient used in the yield plot do not form a line in 3D space. This means instead of predicting yield, we are predicting an ordinal estimate of relative yield. To improve accuracy we will need raw data from the combines.

```python
neighbs = nn.KNeighborsClassifier(n_neighbors=1)
```

```python
neighbs.fit(rgb,range(rgb.shape[0]))
```

```python
X = np.vstack((img[:,:,0].flatten(), img[:,:,1].flatten(), img[:,:,2].flatten())).T
```

```python
y_hat = neighbs.predict(X)
```

```python
plt.hist(y_hat, normed=True, bins=rgb.shape[0])
plt.xticks(range(rgb.shape[0]))
```

Because the color gradient forms a weird line in 3D space, it is difficult to map colors to the average values of the distribution. I skirt around this problem by subtracting 3 (ordinal units of yield) from the higher values.

```python
y_hat[np.argwhere(y_hat > 10)] -= 3
```

Now the yield distribution looks a little more reasonable.

```python
plt.hist(y_hat)
```

```python
yield_vals = y_hat.reshape(img.shape[0],img.shape[1])
```

```python
m = np.mean(yield_vals)
s = np.std(yield_vals)
#Normalize for 0 mean and 1 standard deviation
yield_vals = (yield_vals - m) / s
```

Now we want to make sure that our ordinal yield values roughly correspond to the full color yield map above. Note: lighter == higher yield.

```python
plt.imshow(yield_vals, cmap='bone')
```

For NDVI, we simply look at the red channel. I plot a grayscale again to make sure the values are what we should expect.

```python
xmin,ymin,xmax,ymax = np.floor(zones[1].bounds) + np.array([0,0,1,1])
```

```python
X = np.zeros((len(zones),3))
```

```python
for i,z in enumerate(zones):
    xmin,ymin,xmax,ymax = np.floor(z.bounds) + np.array([0,0,1,1])
    X[i,:] = yield_vals[ymin:ymax,xmin:xmax].mean(), \
             ndvi1[ymin:ymax, xmin:xmax].mean(), \
             ndvi2[ymin:ymax, xmin:xmax].mean()
```

```python
from sklearn.linear_model import LinearRegression, BayesianRidge
from sklearn.cross_validation import train_test_split
from sklearn.metrics import mean_squared_error
```

Split the data into a training set and a test set in order to avoid overconfidence in our model.

```python
trn,tst = train_test_split(X)
```

Create a naive ridge regression model.

```python
mod = BayesianRidge()
```

```python
mod.fit(trn[:,1:],trn[:,0])
```

`mod.score()` is a measure of model fit, $R^2$. It can roughly be interpreted as the amount of variation in the outcome variable explained by the model. 57 % is "good", although the diagnostic is problematic.

```python
mod.score(tst[:,1:],tst[:,0])
```

```python
yhat = mod.predict(tst[:,1:])
```

A better measure of performance is "how close are our predictions to observed values in the test set?" This is plotted in the figure below. the $y=x$ line (in green) would be a perfect fit. Note, we aren't doing great, but there is definitely some signal to be found!

```python
plt.plot(tst[:,0],yhat, 'o')
plt.plot([-2,2],[-2,2])
```

```python
mod.coef_
```

RMSE is another (better) prediction diagnostic. It is in units of the prediction variable. A value of $.4166$ means that on average our predictions are off by .4 of one standard deviation.

```python
np.sqrt(mean_squared_error(tst[:,0], yhat))
```
